```python
#Imports

import numpy as np
from matplotlib import pyplot as plt
from numpy import loadtxt
from statistics import *
```

```python
# Data load - 14k rounds

data_14k_267 = loadtxt('267.data', dtype='int')
data_14k_360 = loadtxt('360.data', dtype='int')

index_14k_267 = np.arange(1,len(data_14k_267)+1,1)
index_14k_360 = np.arange(1,len(data_14k_360)+1,1)

# Data load - 66k rounds
data_66k_267 = loadtxt('267.data', dtype='int')
data_66k_360 = loadtxt('360.data', dtype='int')

index_66k_267 = np.arange(1,len(data_66k_267)+1,1)
index_66k_360 = np.arange(1,len(data_66k_360)+1,1)

data_sorted_66k_267 = np.sort(data_66k_267)
data_sorted_66k_360 = np.sort(data_66k_360)
```

```python
#Some checks and ranges
print("Max 14k: ", max(data_do14k_267))
print("Min 14k: ", min(data_do14k_267))

print("Max 66k: ", max(data_66k_267))
print("Min 66k: ", min(data_66k_267))
```

```python
#Histogram with mean and min values; Maximum is 1750

plt.hist(data_66k_267, bins=300)
plt.axvline(x=statistics.mean(data_66k_267), color = 'r')
plt.axvline(x=min(data_66k_267), color = 'r')
plt.xticks([0, 200, 300, 400, 500, 600, 700, statistics.mean(data_66k_267), min(data_66k_267)])

plt.axis([0, 750, 0, 2000])
plt.title("Histogram with mean and min values; Maximum is 1727; 267 validators")
plt.show()
```

```python
# calculate the proportional values of samples
p = 1. * np.arange(len(data_66k_267)) / (len(data_66k_267) - 1)

# plot the sorted data:
fig = plt.figure()
ax = fig.add_subplot(121)
fig.set_size_inches(12.5, 5.5)

ax.plot(data_sorted_66k_267, p)
ax.axis([0, 1500, 0.95, 1.01])
ax.set_yticks([0.95, 0.96, 0.97, 0.98, 0.99, 0.9996710080043413])
ax.set_xlabel('$time$')
ax.set_ylabel('$percent$')
ax.grid()
plt.axvline(x=600, color='r', linestyle='--')
percent_at_600 = np.interp(600, data_sorted_66k_267, p)
plt.axhline(y=percent_at_600, color='r', linestyle='--')
plt.title("CFD; 99.967% are under 600ms; 267 validators")
plt.show()

```

```python
# calculate the proportional values of samples
p = 1. * np.arange(len(data_66k_267)) / (len(data_66k_267) - 1)

# plot the sorted data:
fig = plt.figure()
ax = fig.add_subplot(121)
fig.set_size_inches(12.5, 5.5)

ax.plot(data_sorted_66k_267, p)
ax.axis([0, 1500, 0.95, 1.01])
ax.set_yticks([0.95, 0.96, 0.97, 0.98, 0.99, 0.9990697133554276])
ax.set_xlabel('$time$')
ax.set_ylabel('$percent$')
ax.grid()
plt.axvline(x=400, color='r', linestyle='--')
percent_at_400 = np.interp(400, data_sorted_66k_267, p)
plt.axhline(y=percent_at_400, color='r', linestyle='--')
plt.title("CFD; 99.907% are under 400ms; 267 validators")
plt.show()

```

```python
plt.figure().set_figwidth(15)

rbp = plt.boxplot([data_66k_267,data_66k_360], notch ='True', vert = 0, whis=(5,95), widths=0.03, flierprops=dict(markerfacecolor='b', marker='D'))

plt.setp(rbp['boxes'], color='k')
plt.setp(rbp['whiskers'], color='k', linestyle='-')
plt.setp(rbp['caps'], color='r', linewidth=3)
plt.setp(rbp["fliers"], markeredgecolor='k', marker='o', markersize=3.0, alpha=0.2)
plt.setp(rbp['medians'], color='#ff0011', linewidth=3)
plt.gca().xaxis.grid(True)
plt.rc('font', size=15) 
plt.xticks([0, 100, 200, 300, 400, 500, 600, 800, 1000, 1200, 1400, 1700, 2000]) 
plt.yticks([1, 2], ['267 validators', '360 validators'])
plt.plot()

```

```python
#What percentage for 600 ms?
percent_at_600 = np.interp(600, data_sorted_66k_267, p)
percent_at_400 = np.interp(400, data_sorted_66k_267, p)

print(percent_at_600, "% are under 600ms")
print(percent_at_400, "% are under 400ms")
```
